```python
import matplotlib.pyplot as plt

fig=plt.figure(figsize = (12,8))
ax1=fig.add_subplot(111, label="1", frame_on=True)
ax2=fig.add_subplot(111, label="2", frame_on=False)
# ax3=fig.add_subplot(111, label="3", frame_on=False)

ax1.plot(w1['TEMP(C)'],(w1['HGHT(m)']/1000), color="C0")
ax1.set_xlabel("Temperature(C)", color="C0")
ax1.set_ylabel("Altitude(KM)", color="C0")
ax1.tick_params(axis='x', colors="C0")
ax1.tick_params(axis='y', colors="C0")

ax2.plot(w1['RELH(%)'],(w1['PRES(hPa)']), color="C1")
ax2.xaxis.tick_top()
ax2.yaxis.tick_right()
ax2.set_xlabel('Relative Humidity(%)', color="C1") 
ax2.set_ylabel('Pressure (hPa)', color="C1")       
ax2.xaxis.set_label_position('top') 
ax2.yaxis.set_label_position('right') 
ax2.tick_params(axis='x', colors="C1")
ax2.tick_params(axis='y', colors="C1")

ax1.spines['bottom'].set_color("C0")
ax1.spines['top'].set_color('C1')
ax1.spines['right'].set_color('C1')
ax1.spines['left'].set_color("C0")


plt.show()
```
